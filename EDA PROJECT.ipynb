```python
#Performing statistical test on SalePrice on the basis of overall quality
```

```python
price_a=df[df['OverallQual']<=5]['SalePrice']
price_b=df[df['OverallQual']>=5]['SalePrice']
```

```python
from statsmodels.stats.weightstats import ztest
```

```python
z_score,p_val=ztest(price_a,price_b)
if p_val>0.05:
    print('No:is True ( There is no difference in price on the basis of Overall Qual)')
else:
     print('Yes:is True ( The Sales Price is different because of Overall Qual)')
```
